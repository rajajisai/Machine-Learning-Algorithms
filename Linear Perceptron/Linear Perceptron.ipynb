```python
#Reading Dataset
def processDataset(file):
    dataset = pd.read_csv(file)
    X=dataset.iloc[:, :].values
    #Shuffling dataset
    shuffle_X = dataset.sample(frac=1)
    X=shuffle_X
    #Append ones to dataset in the beginning
    shuffle_X=np.append(np.ones((np.shape(shuffle_X)[0],1),float),shuffle_X,1)
    #------------Convert target attribute to -1 if its a 0 originally-----------#
    for j in shuffle_X:
        if(j[-1]==0):
            j[-1]=-1

    #Splitting dataset
    train_size = int(0.7 * len(X))
    X_train= np.array(shuffle_X[:train_size])
    X_test= np.array(shuffle_X[train_size:])
    print("X_Train")
    print(X_train)
    print("X_Test")
    print(X_test)
    print("X")
    print(X)
    return X_train,X_test,shuffle_X
    
def getParams():
    W=np.random.uniform(-1,1,(np.shape(X_train)[1]-1,1))
    print("Seed Vector")
    print(W)
    numOfIterations  = 100000
    print("Number of Iterations - ",numOfIterations)
    learning_rate = 1
    return W,numOfIterations,learning_rate

def trainVector(X_train,iterations,learning_rate,X,W):
    #Rotate(Train) the vector in a stochastic fashion
    haltTraining=0
    for i in range(iterations):
        haltTraining=0
        for row in X_train:
            temp=np.reshape(row[:-1],(1,np.shape(X)[1]-1))
            if(np.dot(temp,W)*row[-1]<=0):
                W=W+np.transpose(temp)*row[-1]
                haltTraining=1
                break
        if haltTraining== 0:
            break
            
    print("Weights of Perceptron :\n" ,W)
    return W


def accuracy(X_test,W,X):
    test=np.matmul(X_test[:,:-1],W)
    test=np.sign(test)
    cnt=0
    for i in range(np.shape(X_test)[0]):
        if(X_test[i][np.shape(X)[1]-1]==test[i]):
            cnt+=1;

    print("Accuracy of perceptron",(np.sum(test == X_test[:, -1:],axis=0)/np.shape(X_test)[0]) )
```

```python
print("Dataset 1")
X_train,X_test,X=processDataset("dataset_LP_1.csv")
W,numOfIterations,learning_rate=getParams()
W=trainVector(X_train,numOfIterations,learning_rate,X,W)
accuracy(X_test,W,X)

print("Dataset 2")
X_train,X_test,X=processDataset("dataset_LP_2.csv")
W,numOfIterations,learning_rate=getParams()
W=trainVector(X_train,numOfIterations,learning_rate,X,W)
accuracy(X_test,W,X)
```
